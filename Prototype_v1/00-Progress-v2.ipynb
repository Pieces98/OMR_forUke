```python
vline_opt = ['red', 'darkorange', 'gold', 'greenyellow', 'aquamarine', 'skyblue']        #Color list for cmap

#Declare variable to save calculation results
hline_infos = [{'y':None, 'thicknesses':None} for i in range(len(main_components))]
main_components_staff_removed = [None for i in range(len(main_components))]

#Visualize staff removal progress
fig = plt.figure(figsize=(40,20))
for i, component in enumerate(main_components):
    #Get image pixel histogram
    hist_image, hist_data, grid, grid_idx = fs.get_pixel_histogram(component, grid=[0.3, 0.5, 0.6, 0.7, 0.8, 0.9])
    #Get horizontal line position and thickness from clear area
    hlines, thicknesses = fs.get_staff_lines(hist_image[:,grid_idx[-2]])
    #Remove horizontal lines and save
    component_staff_removed = fs.remove_hlines(component, hlines, thicknesses)
    main_components_staff_removed[i] = component_staff_removed
    
    #Print and save
    print(f'Component[{i:02d}] hline info : {len(hlines)} horizontal lines at y={hlines} with thickness={thicknesses}')
    hline_infos[i]['y'] = hlines
    hline_infos[i]['thicknesses'] = thicknesses
    
    #Show results
    ax = plt.subplot2grid((len(main_components), 5), (i, 0), colspan=2)
    ax.imshow(main_components[i], cmap='gray')
    ax.set_anchor('W')
    
    ax = plt.subplot2grid((len(main_components), 5), (i, 2))
    hist_image2 = cv2.resize(hist_image, dsize=(0, 0), fx=0.5, fy=1.0, interpolation=cv2.INTER_LINEAR)
    ax.imshow(hist_image2, cmap='gray')
    for j, val in enumerate(grid_idx):
        ax.axvline(x=val/2, color=vline_opt[j], label=f'~{100*grid[j]:.1f}%', linewidth=3)
    ax.legend(fontsize='xx-large')
    ax.set_anchor('W')
    
    ax = plt.subplot2grid((len(main_components), 5), (i, 3), colspan=2)
    ax.imshow(component_staff_removed, cmap='gray')
    ax.set_anchor('W')

fig.set_tight_layout(True)
plt.show() 
```

```python
#Classifying symbols
cleves = ['treble', 'bass']
notes = ['quarter_note', 'half_note', 'whole_note']
rests = ['eighth_rest', 'quarter_rest', 'half_rest', 'whole_rest']

def show_templates(temp_dict):
    max_num_type = 0
    for v in temp_dict.values():
        max_num_type = max_num_type if len(v)<max_num_type else len(v)
    num_key = len(temp_dict.keys())
    
    fig = plt.figure(figsize=(2*max_num_type, 2*num_key))
    for i, (k, v) in enumerate(temp_dict.items()):
        for j, img in enumerate(v):
            ax = plt.subplot2grid((num_key, max_num_type), (i, j))
            ax.imshow(img, cmap='gray')
            ax.set_title(f'{k}-{j:02d}')
    fig.set_tight_layout(True)
    plt.show()    
```

```python
#show symbols - cleves
clef_templates = {k:pth.template[k] for k in cleves}
show_templates(clef_templates)
```

```python
#show symbols - notes
note_templates = {k:pth.template[k] for k in notes}
show_templates(note_templates)
```

```python
clist = ['lightcoral', 'chocolate', 'skyblue']
total_time = 0

#Visualize object detection progress
fig = plt.figure(figsize=(30, 5*len(main_components)))
for i, (component_origin, component_removed) in enumerate(zip(main_components, main_components_staff_removed)):   
    ts = time.time()
    print('\n'+f'>>Target : Component : {i:02d}  '+'-='*20)
    
    ax = plt.subplot2grid((len(main_components), 2), (i, 0))
    ax.imshow(main_components[i], cmap='gray')
    ax.set_anchor('W')
    
    ax = plt.subplot2grid((len(main_components), 2), (i, 1))
    ax.imshow(component_origin, cmap='gray')
    
    #Find clef - treble
    #Clef search proceeds from original component image(Given template have staffs)
    print('>>Find Clef - Treble')
    score_t, scale_t, idx_t, shape_t, template_t = fs.find_best_match_v2(component_origin, pth.template["treble"], np.arange(0.5, 0.9, 0.03), search_area_ratio=3)
    print(f'Treble :: pos : (x, y) = {idx_t[::-1]}, shape : (w, h)= {shape_t[::-1]}, best scale : {scale_t:.2f}, best score : {score_t:.2f}')
    ax.add_patch(patches.Rectangle(idx_t[::-1], *shape_t[::-1], edgecolor='gold', fill=False, linewidth=3, label='Treble'))
    
    #Draw staffs 1
    clef_y_from, clef_y_to = idx_t[0]-0.3*shape_t[0], idx_t[0]+1.3*shape_t[0]
    idxs = np.array(hline_infos[i]['y'])
    picks = np.where((clef_y_from <= idxs) & (idxs <= clef_y_to))
    for y in idxs[picks]:
        ax.axhline(y=y, color='gold', linewidth=1.5, linestyle='--')
    
    
    #Next clef search scale depends on treble scale
    scale_list = np.arange(max(0.2, scale_t-0.25), scale_t+0.25, 0.03)
    print(f'Update scale search domain : [min, max] = [{np.min(scale_list):.2f}, {np.max(scale_list):.2f}]')
    
    #Find clef - bass
    print('>>Find Clef - Bass')
    score_b, scale_b, idx_b, shape_b, template_b = fs.find_best_match_v2(component_origin, pth.template["bass"], scale_list, search_area_ratio=3)
    print(f'Bass :: pos : (x, y) = {idx_b[::-1]}, shape : (w, h)= {shape_b[::-1]}, best scale : {scale_b:.2f}, best score : {score_b:.2f}')
    ax.add_patch(patches.Rectangle(idx_b[::-1], *shape_b[::-1], edgecolor='limegreen', fill=False, linewidth=3, label='Bass'))
    
    #Draw staffs 2
    clef_y_from, clef_y_to = idx_b[0]-0.3*shape_b[0], idx_b[0]+1.3*shape_b[0]
    idxs = np.array(hline_infos[i]['y'])
    picks = np.where((clef_y_from <= idxs) & (idxs <= clef_y_to))
    for y in idxs[picks]:
        ax.axhline(y=y, color='limegreen', linewidth=1.5, linestyle='--')
    
    
    #Find Notes
    note_matches = {k:[] for k in note_templates.keys()}
    for j, (note_name, template_list) in enumerate(note_templates.items()):
        #Find notes according to classification
        print(f'>>Find Notes - {note_name}')
        match_infos = fs.find_high_matches_v2(component_removed, template_list, scale_list[::-1], 0.6)
        print(f'Search result for {note_name}...')
        
        total_obj = 0
        for k, info in enumerate(match_infos):
            num_before = info['num']
            #Applying NMS functions to remove overlaped objects
            info = fs.non_maximum_suppression(info, threshold=0.3)
            print(f"Template{k:02d} : {num_before}->{info['num']} objects, scale : {info['scale']:.2f}, shape : {info['shape']}, key score : {info['key_score']:.2f}")
            note_matches[note_name].append(info)
            
            if info['num'] == 0:
                continue
            for y, x in zip(*info['location']):
                ax.add_patch(patches.Rectangle((x, y), *info['shape'][::-1],  edgecolor=clist[j], fill=False, linewidth=2))
            total_obj += info['num']
        ax.add_patch(patches.Rectangle((0,0), 0, 0, fill=False, edgecolor=clist[j], linewidth=2, label=f'{note_name}:{total_obj:02d} obj'))

    ax.legend(loc='lower center', fontsize='large', bbox_to_anchor=(0.5, -0.15), ncol=5)
    ax.axis('off'); ax.set_anchor('W')
    
    tf = time.time()
    print(f'*Runtime : {tf-ts:.2f} sec')
    total_time += tf-ts
    
print(f'###Total Runtime for {len(main_components)} components : {total_time:.2f} sec###')

fig.set_tight_layout(True)
plt.show() 
```
